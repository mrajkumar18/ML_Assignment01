#### `Fill the missing values in the column 'Outlet_Size' by Medium using loc`

---

```python
data.isna().sum() 
```

```python
data.Outlet_Size.value_counts()
```

```python
data.Outlet_Size.dtypes
```

```python
# WRITE YOUR CODE HERE - QUESTION 4
# Fill the missing values in the column 'Outlet_Size' by Medium using loc

data.loc[(data.Outlet_Size.isna() == True) , 'Outlet_Size'] = 'Medium'
```

```python
data.isna().sum()
```

```python
data.Item_Weight
```

```python
data.Item_Weight.mean()
```

```python
# Missing vlaue in Item_Weight replaced with Item_Weight
data.loc[(data.Item_Weight.isna() == True) , 'Item_Weight'] = (data.Item_Weight.mean())
```
